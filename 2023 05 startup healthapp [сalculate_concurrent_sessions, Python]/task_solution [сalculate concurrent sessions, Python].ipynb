Для того, чтобы рассмотреть получше, избавимся от экстремальных значений. Кроме того, такие значения, т.е. аномально длительные сессии, будут содержать в себе огромное количество коротких сессий, а аномально короткие сессии будут включаться в более длительные, что смажет нам общую картину, если этого не сделать. <br />

<b> Для принятия решений по работе с экстремальными значениями нам не хватает  контекста и информации об активности каждого пользователя, т.к. колонки 'start_edit_time' 	и 'conn_end_time'	отражают только начало и конец сессии, и никак не характеризуют активность юзера. <b />

Поэтому, основываясь лишь на имеющихся данных, для исключения экстремальные значений определим пороговое значение для выбросов. 
В нашем случае, в случае экспоненциального распределения, выбросы находятся в правом хвосте.<br />  Воспользуемся методом межквартильного размаха (IQR).

```python
Q1 = df['duration'].quantile(0.25)
Q3 = df['duration'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR  
# Учитывая особенности правостороннего экспоненциального распределения, где все значения положительны, смысла в расчёте нижнего пограничного значения смысла нет.
```

```python
print(
"* Нижнее пороговое значение: ", lower_bound, '\n',
"* Верхнее пороговое значение: ", upper_bound,
sep = '')
```

Учитывая особенности правостороннего экспоненциального распределения, где все значения положительны, так как duration (длительность) не может быть < 0, смысла в расчёте нижнего пограничного значения смысла нет. <br>
Но на мой взгляд, необходимо избавиться от слишком маленьких значений. Поэтому я решил также убрать записи с длительностью сессии меньше значения 2.5% перцентиля (~30 сек). Данное значение должно определяться c учётом контекста. Т.к. информации о контексте нет, оно складывается лишь из моего представления о экстримально низких значениях. <br /> По моему мнению, это исключит пользователей, случайно или ошибочно зашедших в режим редактирования документов и в последуещем, избавит нас от лишнего шума в данных.

```python
lower_bound = df['duration'].quantile(0.025)
lower_bound
```

Видим всплеск сессий  длительностью  30 - 35 мин. и нисходящее утолщение хвоста распределения после, что предположительно связано с настройками автоматического отключения (пользователь не закрывает документ сам, через полчаса неактивности он автоматически отключается, что фиксируется в conn_end_time). <br />
Причем, данный всплеск, выпал на значения 30 - 35 мин. не спроста, т.к. если мы вычтем от таких значений 30 мин., то как раз и получим значения из самых частовстречаемых по длительности сессий.

Есть предположение, что если мы расчитаем для каждой сессии количество одновременных сессий, то переменные 'длительность сессии' и 'количество одновременных сессий' будут иметь положительную линейная связь. 

Так посчитаем же количество пересечений для каждой из сессий.

```python
# Создадим новый датафрейм df_test на основании df_new.
df_test = df_new[['user', 'start_edit_time', 'conn_end_time', 'duration']].copy()
```

```python
# Для удобства переименуем колонки следующим образом
df_test = df_test.rename(columns={'user': 'user_id', 'start_edit_time': 'login', 'conn_end_time': 'logout'})
df_test.head(2)
```

```python
# Добавим колонку с индивидуальным индентификатором каждой сессии, 
# который будет складываться со значений колонки user и индекса строки.
df_test['sid'] = df_test.user_id.astype(str) + "-" + df_test.index.astype(str)
df_test.head(2)
```

```python
# Преобразуем датафрейм из широкого формата в длинный формат.
df_chrono = pd.melt(df_test.set_index('sid'), value_vars=['login', 'logout'], ignore_index=False)
df_chrono = df_chrono.sort_values(by='value').reset_index()
df_chrono
```

```python
# 1. Создадим пустой словарь simul_track для отслеживания количества пересечений для каждой сессии
simul_track = {}

# 2. Создадим словарь results для сохранения результатов
results = {"sid": [], "simul": []}

# 3. Проходимся по каждой строке датафрейма df_chrono
for i, row in df_chrono.iterrows():
    # 4. Если текущая строка представляет собой 'login'
    if row.variable == 'login':
        # 5. Увеличиваем количество одновременных подключений для каждого sid в simul_track
        for sid in simul_track:
            simul_track[sid] += 1

    # 6. Проверяем, если sid текущей строки не находится в simul_track
    if row.sid not in simul_track:
        # 7. Добавляем новый sid в simul_track и устанавливаем количество пересечений равным текущей длине simul_track
        simul_track[row.sid] = len(simul_track)
    else:
        # 8. Если sid уже присутствует в simul_track, извлекаем его (количество пересечений) и добавляем в результаты
        count = simul_track.pop(row.sid)
        results['simul'].append(count)
        results['sid'].append(row.sid)
```

```python
# На основании полученного словоря создадим новый датафрейм df_results
df_results = pd.DataFrame(results,  columns=['sid','simul'])
df_results.head(2)
```

```python
# Объеденим датафреймы df_test и f_results по индетификатору сессии, оставив лишь пересекающиеся записи.
df_res=pd.merge(df_test, df_results, on='sid', how='inner').sort_values(by='simul', ascending=False)
df_res
```

Рассчитав количество пересечений для каждой сесссии, теперь мы можем с помощью коэффициента Пирсона оценить линейную связь между  переменными 'длительность сессии'(duration) и 'количество одновременных сессий'(simul).

Это значение говорит о том, что между этими двумя переменными есть достаточно сильная положительная линейная связь.  Также, её можно увидеть на матрице диаграмм рассеяния. <b> Т.е. мы можем говорить о  значительной зависимости двух переменных. <br />

Значение коэффициента корреляции 0.78 близко к 1, что указывает на то, что с увеличением длительности сессии обычно увеличивается и количество одновременных сессий.

Корреляция не означает причинно-следственную связь между переменными, а лишь показывает наличие связи между ними.

И наконец посчитаем количество одновременных сессий. Под одновременностью будем понимать количество сессий в минуту.

Для этого:

```python
# Построим новый датафрейм с двумя колонками: 'time' c разбивкой исследуемого интервала по минуте 
# и колонкой  simultaneous_sessions, которую для начала заполним нулями.
start_time = '2022-12-04 08:07:25'
end_time = '2022-12-07 01:30:50'
time_range = pd.date_range(start=start_time, end=end_time, freq='1min') # Создадим диапазон дат с разбивкой по минуте
df_simultaneous_sessions = pd.DataFrame({'time': time_range, 'simultaneous_sessions': 0})
df_simultaneous_sessions.head()
```

```python
# Создадим цикл, который пройдётся по исходному датафрейму df 
# и в случае попадения значения колонки time (df_simultaneous_sessions) в диапозон между start_edit_time и conn_end_time
# увеличит значение simultaneous_sessions на 1.

# Заполнение колонки simultaneous_sessions значениями из исходного датафрейма df
for index, row in df.iterrows():
    start_time = row['start_edit_time']
    end_time = row['conn_end_time']
    
    # Увеличение значения simultaneous_sessions для каждой минуты в интервале
    mask = (df_simultaneous_sessions['time'] >= start_time) & (df_simultaneous_sessions['time'] <= end_time)
    df_simultaneous_sessions.loc[mask, 'simultaneous_sessions'] += 1

df_simultaneous_sessions.head()
```
